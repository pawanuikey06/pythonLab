```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =pd.read_csv('homeprices.csv')
```

```python
df
```

```python
%matplotlib inline  
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US $)')
plt.scatter(df.area, df.price, color='red', marker='o')  # Use 'o' for circular markers
plt.show()
```

```python
model =linear_model.LinearRegression()

model.fit(df[['area']],df.price)
```

```python
X_new = np.array([3300]).reshape(-1, 1)  # Reshape to 2D array (1 sample, 1 feature)
prediction = model.predict(X_new)
print(prediction)
```
